```python
# Data cleaning

# option 1: drop or get rid of null values

df.isnull().sum()
```

```python
df.dropna()  # any row with null value has been removed
```

```python
# i have not place inplace = True ,so my original df is untouched
df
```

```python
 df.dropna(how = "any")    # its same; it remove rows with null values # any row that have any null value
```

```python
df.dropna(how = "all")   # if all the values in any row are null, we are going to drop that row 
```

```python
# option 2: to fill those missing values

# its important
"""
fillna()

methods
1. forwardfill - top to bottom
2. backwardfill - bottom to top

"""

# df.fillna(0)  # noob
df["Age"].fillna(df["Age"].mean())    # missing value in age has been replaced by mean
```

```python
df["Salary"].fillna(df["Salary"].median())   # missing value in salary has been replaced by median
```

```python
# 1. forward fill

df["Age"].fillna(method = "ffill")
```

```python
# 2. backward fill

df["Age"].fillna(method = "bfill")
```

```python
# this may give error if our first and last value is null
```

```python
# replace name with other name

df["Name"].replace("Charlie", "Rose")
```

```python
df["Name"] = df["Name"].replace("Charlie","Rose")  # to replace permanently
```

```python
df
```

```python
# deal with duplicate values

# we have inbuilt function

df_dup = df[df.duplicated()]
df_dup
```

```python
# by default parameter is (keep = "first")

df_dup = df[df.duplicated(keep = "first")]   # top to bottom
df_dup
```

```python
# when we say (keep = "last"); our first element is duplicated

df_dup = df[df.duplicated(keep = "last")]    # bottom to top
df_dup
```

```python
# i prefer my second record to be deleted

df = df.drop_duplicates()
df
```

```python
# how to deal with invalid values

# to do changes in the column of the particular dataset
# lamda - python

df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x/10 if x > 650000 else x)
df
```

```python

"""
this is how to split our data set

name = "alice_fernades"
df[["first_name","last_name"]] = df["name"].str.split("_")
"""
```

```python
# multiply all the ages with 2

# apply and lamda

def multiplying_age(x):
    return x * 2

df["Age"] = df["Age"].apply(multiplying_age)
df
```

```python
# return to original value

df["Age"] = df["Age"].apply(lambda x: x/2)
df
```

```python
# Joins and Merges

"""
like in sets

left join - A-B  # only A
right join - B-A  # only B
outer join - AUB  # A union B
inner join - AnB  # A intersection B

merge is like inner join

"""

department_info = {
    "Department" : ["HR","IT","Finance"],
    "Location" : ["New York","San Francisco","Chicago"],
    "Manager" : ["Laura","Steve","Nina"]
}

df2 = pd.DataFrame(department_info)
df2


```
